# 一维梯度下降可视化

函数：$y = x^2$,设x的初始值为12，学习率为0.2，使用梯度下降对x迭代10次，绘制x轨迹

```python
import matplotlib.pyplot as plt
import numpy as np

def get_trace(learning_rate):
    x = 10
    trace = [x,]
    for epoch in range(10):
        x -= learning_rate * 2 * x
        trace.append(x)
    return trace

def plot_trace(trace):
    n = max(abs(min(trace)),abs(max(trace)))
    plt.Figure(figsize=(6,6))
    x = np.arange(-n,n,0.1)
    plt.plot(x,[i*i for i in x])
    plt.plot(trace, [i * i for i in trace], 'r-o')
    plt.show()
```

```python
# learning_rate = 0.2
plot_trace(get_trace(0.2))
```

```python
# learning_rate = 1
plot_trace(get_trace(1.1))
```

# 多维梯度下降可视化

函数：$f(x_1,x_2) = {x_1}^2 + 2{x_2}^2$,从初始位置(-5,-2)对自变量$(x_1,x_2)$的20次迭代轨迹进行可视化

```python
def fun(x1,x2):
    return x1**2 + 2 * (x2 ** 2)

def get_gradient(x1,x2,learning_rate):
    return x1 - 2*x1*learning_rate, x2 - 4*x2*learning_rate

def train(learning_rate):
    x1 = -5.0
    x2 = -2.0
    result = [(x1,x2)]
    for i in range(20):
        x1,x2 = get_gradient(x1,x2,learning_rate)
        result.append((x1,x2))
    return result

def plot_trace_2d(trace):
    plt.Figure(figsize=(7,7))
    plt.plot(*zip(*trace), 'r-o')
    x1,x2 = np.meshgrid(np.arange(-5.5,1.0,0.1),np.arange(-3.0,1.0,0.1))
    plt.contour(x1,x2,fun(x1,x2),colors='#1f77b4')
    plt.show()
```

```python
plot_trace_2d(train(0.1))
```
